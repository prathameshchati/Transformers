#### Positional embeddings visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def positional_encodings(position, d_model):
    angle_rads = np.arange(position)[:, np.newaxis] / np.power(10000, (2 * (np.arange(d_model)[np.newaxis, :] // 2)) / np.float32(d_model))
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
    return angle_rads
```

```python
# generate
position=50  
d_model=128

pos_encoding=positional_encodings(position, d_model)

plt.figure(figsize=(12, 8))
plt.imshow(pos_encoding, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.xlabel('i')
plt.ylabel('Position')
plt.title('Position Embeddings')
plt.show()
```
